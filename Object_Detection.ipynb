```python
import pandas as pd
import numpy as np

# Function to detect outliers using Z-score
def detect_outliers_zscore(data):
    threshold = 3
    outliers = []
    for column in data.columns:
        mean = np.mean(data[column])
        std = np.std(data[column])
        for i, value in enumerate(data[column]):
            z_score = (value - mean) / std
            if np.abs(z_score) > threshold:
                outliers.append((i, column))
    return outliers

outliers_check_df = df.iloc[:, 1:]
outliers = detect_outliers_zscore(outliers_check_df)

# Print outliers and their corresponding values
print("Outliers detected using Z-score method:")
for outlier in outliers:
    index, column = outlier
    print(f"Outlier in {column} at index {index}: {df.loc[index, column]}")


valid_outliers = [(index, column) for index, column in outliers if df.loc[index, 'xmin'] >= df.loc[index, 'xmax'] or df.loc[index, 'ymin'] >= df.loc[index, 'ymax']]
print("\nValid outliers:")
indexes_of_valid_outliers = set()
for outlier in valid_outliers:
    index, column = outlier
    print(f"Outlier in {column} at index {index}: {df.loc[index, column]}")
    indexes_of_valid_outliers.add(index)
```

```python
#Getting the path of all the xml files
from glob import glob
xml_path = glob('./images/*.xml')
xml_path[:5]
```

```python
#Extracting the height and width of number plate from the xml files
import xml.etree.ElementTree as xet
labels_dict = dict(height = [] , width = [])
for filePath in xml_path: 
    info = xet.parse(filePath)
    root = info.getroot()
    labels_info = root.find('size')
    height = int(labels_info.find('height').text)
    width = int(labels_info.find('width').text)
    labels_dict['height'].append(height)
    labels_dict['width'].append(width)
```

```python
#converting Dictionary to Dataframe
height_width_info = pd.DataFrame(labels_dict)
display(height_width_info.head())
```
